## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
The different relationships that can be found in a SQL database are:
 --> one-to-one:
     As the name suggests a single entry from one table A will connect or have "a relationship" with a single entry from another table B. 
 --> one-to-many:
     Here one entry from a table A can link up to more then one entry in table B. However, one entry from table B can only have a relationship with one entry from table A. 
 --> many-to-many:
     many entries in one table A share relationships with multiple entries of another table B.
